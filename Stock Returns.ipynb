```python
df = pdr.get_data_yahoo(stock,start_date,now)
```

```python
ema_used = [5,10,50]

for i in ema_used:
    ema_string = 'ema'+str(i)
    df[ema_string] = round(df.iloc[:,4].ewm(span=i).mean(),2)
    
df.tail()
    
```

```python
pos = 0
pct_change =[]
num =0

for i in df.index[:200]:
    cmin= min(df['ema5'][i],df['ema10'][i])
    cmax=df['ema50'][i]
    
    if cmin>cmax:
        print ('red white blue')
        if pos==0:
            pos =1
            buying_price = round(df['Adj Close'][i],2)
            print('buying now at '+str(buying_price))
        
    else:
        print ('blue white red')
        if pos ==1:
            pos =0
            selling_price = round(df['Adj Close'][i],2)
            percentage_change = round(((selling_price/buying_price-1)*100),2)
            pct_change.append(percentage_change)
            print('selling now at '+str(selling_price))
    if num==df['Adj Close'].count()-1 and pos ==1:
            pos =0
            selling_price = round(df['Adj Close'][i],2)
            percentage_change = round(((selling_price/buying_price-1)*100),2)
            pct_change.append(percentage_change)
            print('selling now at '+str(selling_price)+' because last date')
    num+=1
sum(pct_change)
```

```python
gain_count =0
loss_count = 0
total_return =1
gains =0
losses=0

for i in pct_change:
    if i>0:
        gain_count+=1
        gains+=i
        
    elif i<0:
        loss_count+=1
        losses+=i
        
    total_return = total_return*(i+100)/100

total_return = round(((total_return-1)*100),2)

print(pct_change)
print(total_return)
```

```python
if gain_count>0:
    max_gain = str(max(pct_change))
    avg_gain = gains/gain_count
    
elif gain_count ==0:
    avg_gain =0
    max_gain ='nil'
    
if loss_count>0:
    max_loss=str(min(pct_change))
    avg_loss = losses/loss_count
    ratio = (-avg_gain/avg_loss)

elif loss_count==0:
    avg_loss = 0
    max_loss = 'nil'
    ratio = 'Undefined as no losses'

ratio
```

```python
trade_success = gain_count/(gain_count+loss_count)

trade_success

df
```
